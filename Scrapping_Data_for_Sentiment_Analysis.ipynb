##loading libraries

```python
import requests
from bs4 import BeautifulSoup as bs
```

## lets the load the page

```python
# load webpage content
rq = requests.get("https://www.best-poems.net/rumi/a-moment-of-happiness.html")
```

```python
# convert to a beautiful soup object
soup = bs(rq.content)
```

```python
# print out html, prettify just for better understanding
print(soup.prettify())
```

```python
soup.find_all("div")
```

```python
soup.find("div", attrs={"id":"content"})
```

```python
div_content = soup.find("div", attrs={"id":"content"})
```

```python
para = div_content.find("p")
```

```python
print(para)
```

```python
soup.select("div p")
```

```python
get_poem_text = para.get_text()
```
